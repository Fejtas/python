# SQL DML: Lednice

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect(':memory:')
cursor = conn.cursor()
cursor.execute("CREATE TABLE Lednice (Jidlo TEXT, Pocet_kusu INTEGER)")
conn.commit()
```

## 1. INSERT

```python
cursor.execute("INSERT INTO Lednice VALUES ('Jablko', 5)")
conn.commit()
```

## 2. SELECT

```python
display(pd.read_sql_query("SELECT * FROM Lednice", conn))
```

## 3. UPDATE

```python
cursor.execute("UPDATE Lednice SET Pocet_kusu = 3 WHERE Jidlo = 'Jablko'")
conn.commit()
display(pd.read_sql_query("SELECT * FROM Lednice", conn))
```

## 4. DELETE

```python
cursor.execute("DELETE FROM Lednice WHERE Jidlo = 'Jablko'")
conn.commit()
display(pd.read_sql_query("SELECT * FROM Lednice", conn))
```
